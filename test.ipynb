```python
import pandas as pd

# CSV 파일 로드
df = pd.read_csv('/data/2_data_server/nlp-04/lost_technology/original_data/banking/train_10.csv')

# 'class'를 기준으로 그룹화하고, 각 그룹에 대해 'text' 컬럼의 값을 리스트로 변환
result = df.groupby('category')['text'].apply(list).to_dict()
```

```python
total_length = len(result)
part_length = total_length // 8
parts = [list(result.items())[i*part_length : (i+1)*part_length] for i in range(8)]
for i, part in enumerate(parts):
    print(f"Part {i+1}: {part}")
```
